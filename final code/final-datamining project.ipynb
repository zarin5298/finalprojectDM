```python
df = df.sample(n=10000).drop('Weekly.rank', axis=1).reset_index(drop=True)
```

```python
df.drop(['Peak.position','Weeks.on.chart','Week','Date','Features'],axis=1, inplace=True)
```

```python
df.sort_values("Name", inplace = True)
```

```python
features = ['Genre', 'Lyrics']
features
```

```python
df[features] = df[features].replace({',': ''}, regex=True)
```

```python
df['data'] = df['Genre'] + " " + df['Lyrics']
df['data']
```

```python
df['data'].iloc[10]
```

```python
import numpy as np
```

```python
from typing import List, Dict
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(analyzer='word', stop_words='english')
```

```python
lyrics_matrix = tfidf.fit_transform(df['data'])
```

```python
cosine_similarities = cosine_similarity(lyrics_matrix)
```

```python
similarities = {}
```

```python
for i in range(len(cosine_similarities)):
    similar_indices = cosine_similarities[i].argsort()[:-50:-1]
    similarities[df['Name'].iloc[i]] = [(cosine_similarities[i][x], df['Name'][x], df['Artists'][x]) for x in similar_indices][1:]
```

```python
class ContentBasedRecommender:
    def __init__(self, matrix):
        self.matrix_similar = matrix

    def _print_message(self, Name, recom_Name):
        rec_items = len(recom_Name)
        
        print(f'The {rec_items} recommended names for {Name} are:')
        for i in range(rec_items):
            print(f"Number {i+1}:")
            print(f"{recom_Name[i][1]} by {recom_Name[i][2]} with {round(recom_Name[i][0], 3)} similarity score") 
            print("--------------------")
        
    def recommend(self, recommendation):
        # Get song to find recommendations for
        Name = recommendation['Name']
        # Get number of songs to recommend
        number_Names = recommendation['number_Names']
        # Get the number of songs most similars from matrix similarities
        recom_Name = self.matrix_similar[Name][:number_Names]
        # print each item
        self._print_message(Name=Name, recom_Name = recom_Name)
```

```python
recommedations = ContentBasedRecommender(similarities)
```

```python
recommendation = {
    "Name": df['Name'].iloc[1001],
    "number_Names": 10
}
```

```python
recommedations.recommend(recommendation)
```
